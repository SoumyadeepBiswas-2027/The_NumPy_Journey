```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr1 = np.array([[1,2,3],[4,5,6]])
arr2 = np.random.rand(3,3)
arr3 = np.zeros((4,4))
```

```python
np.save('arr1.npy',arr1)
np.save('arr2.npy',arr2)
np.save('arr3.npy',arr3)
```

```python
loaded_arr1 = np.load('arr1.npy')
print(loaded_arr1)
```

```python
from matplotlib import image

# Step 1: Load image as array
img = image.imread("Pictures/dimitri.jpg")

# Step 2: Save it as a .npy file
np.save("Pictures/dimitri.npy", img)

print("Image successfully converted to dimitri.npy")
```

```python
try:
    pic = np.load('Pictures/dimitri.npy')

    #Display
    plt.figure(figsize=(10,5))
    plt.subplot(121)
    plt.imshow(pic)
    plt.title("Numpy picture")
    plt.grid(False)

    dark_pic = 1- pic
    plt.subplot(122)
    plt.imshow(dark_pic)
    plt.title("Numpy dark picture")
    plt.grid(False)
    
except FileNotFoundError:
    print("picture not found")
```
